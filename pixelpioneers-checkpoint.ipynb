```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread(r'C:\Users\Lenovo\OneDrive\Desktop\lab3\added_image_new.jpg')

# Split the image into its color channels (BGR)
b, g, r = cv2.split(image)

# Calculate the histograms of each color channel before normalization
hist_b_before = cv2.calcHist([b], [0], None, [256], [0, 256])
hist_g_before = cv2.calcHist([g], [0], None, [256], [0, 256])
hist_r_before = cv2.calcHist([r], [0], None, [256], [0, 256])

# Plot the histograms of each color channel before normalization
plt.figure(figsize=(10, 6))
plt.subplot(3, 1, 1)
plt.plot(hist_b_before, color='blue')
plt.title('Histogram of Blue Channel (Before Normalization)')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.subplot(3, 1, 2)
plt.plot(hist_g_before, color='green')
plt.title('Histogram of Green Channel (Before Normalization)')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.subplot(3, 1, 3)
plt.plot(hist_r_before, color='red')
plt.title('Histogram of Red Channel (Before Normalization)')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Normalize each color channel to achieve a more balanced color representation
normalized_b = cv2.normalize(b, None, 0, 255, cv2.NORM_MINMAX)
normalized_g = cv2.normalize(g, None, 0, 255, cv2.NORM_MINMAX)
normalized_r = cv2.normalize(r, None, 0, 255, cv2.NORM_MINMAX)

# Calculate the histograms of each color channel after normalization
hist_b_after = cv2.calcHist([normalized_b], [0], None, [256], [0, 256])
hist_g_after = cv2.calcHist([normalized_g], [0], None, [256], [0, 256])
hist_r_after = cv2.calcHist([normalized_r], [0], None, [256], [0, 256])

# Plot the histograms of each color channel after normalization
plt.figure(figsize=(10, 6))
plt.subplot(3, 1, 1)
plt.plot(hist_b_after, color='blue')
plt.title('Histogram of Blue Channel (After Normalization)')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.subplot(3, 1, 2)
plt.plot(hist_g_after, color='green')
plt.title('Histogram of Green Channel (After Normalization)')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.subplot(3, 1, 3)
plt.plot(hist_r_after, color='red')
plt.title('Histogram of Red Channel (After Normalization)')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Merge the normalized color channels back into a single image
normalized_image_4 = cv2.merge([normalized_b, normalized_g, normalized_r])

# Display the normalized image
cv2.imshow('Normalized Image_4', normalized_image_4)

# Save the normalized image
cv2.imwrite('normalized_image_4.jpg', normalized_image_4)

# Wait for a key press and close all OpenCV windows
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def plot_histogram(image, title):
    """Plot histogram of the given image."""
    histogram = cv2.calcHist([image], [0], None, [256], [0, 256])
    plt.figure()
    plt.title(title)
    plt.xlabel("Pixel Value")
    plt.ylabel("Frequency")
    plt.plot(histogram)
    plt.xlim([0, 256])

def contrast_stretching(image):
    """Apply contrast stretching to the given image."""
    # Compute the minimum and maximum intensity values from the image.
    minval = np.min(image)
    maxval = np.max(image)
    print ('Existing min, max values :', minval, ', ', maxval) 
    dmin = 0
    dmax = 255
    print ('Desired min, max values :', dmin, ', ', dmax) 
    # Apply contrast stretching.
    dgap = dmax - dmin
    stretched_image = dmin + (image - minval) * (dgap / (maxval - minval))
    return stretched_image.astype(np.uint8)

# Load the image
image_path = r'C:\Users\Lenovo\OneDrive\Desktop\lab3\normalized_image_4.jpg'  # Update this path
original_image = cv2.imread(image_path)

# Split the image into its color channels (BGR)
b, g, r = cv2.split(original_image)

# Apply contrast stretching to each color channel separately
stretched_b = contrast_stretching(b)
stretched_g = contrast_stretching(g)
stretched_r = contrast_stretching(r)

# Merge the stretched color channels back into a single image
stretched_image_2 = cv2.merge([stretched_b, stretched_g, stretched_r])

# Plot histograms
plot_histogram(b, "Histogram for Blue Channel")
plot_histogram(g, "Histogram for Green Channel")
plot_histogram(r, "Histogram for Red Channel")
plot_histogram(stretched_b, "Histogram for Stretched Blue Channel")
plot_histogram(stretched_g, "Histogram for Stretched Green Channel")
plot_histogram(stretched_r, "Histogram for Stretched Red Channel")

# Display the original and stretched images
cv2.imshow("Original Image", original_image)
cv2.imshow("Contrast Stretched Image", stretched_image_2)
# Save the output image
cv2.imwrite('stretched_image_2.jpg', stretched_image_2)

plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
